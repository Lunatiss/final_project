### Get information for T-shirts category using Selenium & Beautiful library

```python
# Inicializar WebDriver
driver = webdriver.Chrome()

# Abrir la página de Zara
driver.get('https://www.zara.com/es/es/mujer-camisetas-l1362.html?v1=2352955&regionGroupId=110')

driver.maximize_window()
```

```python
# Scrolling the web page
height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(5)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if height == new_height:
        break
    height = new_height
```

```python
product_links = []

# Getting the product elements
page_product_links = driver.find_elements(By.XPATH, '//div[@class="product-grid-product__figure"]/a')

# Getting the product links
for product in page_product_links:
    product_link = product.get_attribute('href')
    product_links.append(product_link)
```

```python
product_links
```

```python
len(product_links)
```

```python
def extract_info_ok(link):  # Nota: corrige el nombre de la función de 'extact_info' a 'extract_info'
    driver = webdriver.Chrome()
    driver.get(link)
    
    time.sleep(5)

    # Obtener el contenido de la página actual
    page_source = driver.page_source
        
    # Analizar el contenido de la página con BeautifulSoup
    profile = BeautifulSoup(page_source, 'html.parser')
    
    # Esperar 5 segundos antes de extraer la información (15 segundos pueden ser innecesarios)
    #time.sleep(5)
        
    # Realizar un diccionario para que la información se vaya guardando.
    clothes_dict = {}
                
    # Extraer información del perfil
    product_name = profile.find("h1", class_='product-detail-info__header-name')
    clothes_dict["product_name"] = product_name.text if product_name else None
        
    price = profile.find("span", class_='money-amount__main')
    clothes_dict["price"] = price.text if price else None
    
    #time.sleep(5)
    item_description = [item.get_text(strip=True) for item in profile.find_all("span", class_='structured-component-text zds-paragraph-m')]
    clothes_dict['description'] = item_description
    #print(item_description)

    # Imprime el diccionario para depuración
    #print(clothes_dict)
    
    driver.quit
    
    return clothes_dict
```

```python
cloth = [extract_info_ok(link) for link in product_links[:5]]

# Make a DataFrame from a dict.
df = pd.DataFrame(cloth)

# Print DataFrame
df
```
